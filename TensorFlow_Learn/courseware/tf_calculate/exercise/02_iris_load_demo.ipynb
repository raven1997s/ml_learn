```python
#从sklearn 中获取鸢尾花数据集
from sklearn import datasets
from pandas import DataFrame
import pandas as pd

# 返回iris数据集所有输入特征
x_data = datasets.load_iris().data

# 返回iris数据集所有标签
y_data = datasets.load_iris().target
print(f'x_data from sklearn.datasets.load_iris():\n{x_data}')
print(f'y_data from sklearn.datasets.load_iris():\n{y_data}')
```

```python

# 为表格增加行索引（左侧）和列标签（上方）
x_data = DataFrame(x_data, columns=['花萼长度', '花萼宽度', '花瓣长度', '花瓣宽度'])


# 设置列名对齐
pd.set_option('display.unicode.east_asian_width', True)

print(x_data)
```

```python
# 新加一列，列标签‘类别’ 数据为y_data
x_data['类别'] = y_data

print(x_data)
```

```python
from tensorflow import keras
import numpy as np

# one bedroom 50k +50k  two bedroom 50k + 100k
# scale 10k  to 1 
# y = 5x + 5

model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')

x = np.array([1, 2 ,3 ,4, 5], dtype=float)
y = np.array([10, 15,20,25,30], dtype=float)

model.fit(x, y, epochs=1000)
```

```python
price = model.predict(np.array([7]));
print(price)
weights,biases = model.layers[0].get_weights();
print(f"权重（Weights）: {weights}")
print(f"偏置（Biases）: {biases}")

actual_price =  price[0][0] * 10000
print(f"实际价格: {actual_price}")
```

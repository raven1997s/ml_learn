```python
from torch.utils.tensorboard import SummaryWriter
    
from PIL import Image
import numpy as np 
writer = SummaryWriter('logs')

# writer.add_image()

image_path ='./practice_data/train/ants_image/0013035.jpg'
# image_path ='./practice_data/train/ants_image/5650366_e22b7e1065.jpg'
img = Image.open(image_path)

print(f'img : {img}')
print(f'img type : {type(img)}')

img_array = np.array(img)
print(f'img_array : {img_array}')
print(f'img_array type : {type(img_array)}')
print(f'img_array shape : {img_array.shape}')


writer.add_image('test', img_array, 2, dataformats='HWC') 

# writer.add_scalar()
# y = x 
for i in range(100):
    writer.add_scalar('y=x', i, i)

# y = 2x
for i in range(100):
    writer.add_scalar('y=2x', 2*i, i)

writer.close()
```

```python
import torch
import torch.nn as nn
import torchvision

# dataloader
from torch.utils.data import DataLoader

dataset = torchvision.datasets.CIFAR10(
    root="./dataset",
    transform=torchvision.transforms.ToTensor(),
    train=False,
    download=True,
)

dataloader = DataLoader(dataset=dataset, batch_size=64, shuffle=True,drop_last=True)


class MyModule(nn.Module):
    def __init__(self):
        super(MyModule, self).__init__()
        # 线性层，输入为64*3*32*32，输出为64*10
        # 这里线性层的作用为将输入的多维数据压缩成一维数据
        # 64*3*32*32 = 196608
        # 64表示batch_size，3表示通道数，32表示高度，32表示宽度
        self.linear1 = nn.Linear(196608, 10)

    def forward(self, x):
        x = self.linear1(x)
        return x


myModule = MyModule()


for data in dataloader:
    imgs, targets = data
    print(f"imgs.shape: {imgs.shape}")
    # 将多维数据压缩成一维数据，便于后续全连接层处理
    # 这里的第一个1表示batch_size，第二个1表示通道数，第三个1表示高度，最后一个-1表示宽度
    # output = torch.reshape(imgs, (1, 1, 1, -1))
    # print(f"output.shape: {output.shape}")

    # troch.flatten(imgs)
    # 将数据压缩成一维数据，便于后续全连接层处理
    # flatten 和 reshape的区别
    # flatten会将数据压缩成一维数据，reshape不会
    output = torch.flatten(imgs)
    print(f"output.shape: {output.shape}")

    # 将数据传入模型

    output = myModule(output)
    print(f"output.shape: {output.shape}")
```

```python
import json
import os
import requests

# 读取 JSON 文件
with open('coin.txt', 'r', encoding='utf-8') as file:
    data = json.load(file)

# 创建保存图片的文件夹
output_dir = 'coin'
os.makedirs(output_dir, exist_ok=True)

# 遍历 JSON 数据中的每个 key
for key, value in data.items():
    # 获取 logo URL
    logo_url = value[0].get('logo')
    if logo_url:
        # 下载并保存图片
        try:
            response = requests.get(logo_url)
            response.raise_for_status()  # 检查请求是否成功

            # 确定图片保存路径
            file_path = os.path.join(output_dir, f'{key}_ICON.png')

            # 将图片写入文件
            with open(file_path, 'wb') as img_file:
                img_file.write(response.content)

            print(f"图片 {file_path} 下载成功！")

        except requests.RequestException as e:
            print(f"下载 {logo_url} 时出错: {e}")
```

```python
# 读取 bank.txt 文件内容
with open("bank.txt", "r") as file:
    bank_names = file.read().splitlines()

# 打开 sql.sql 文件以写入模式
with open("sql.sql", "w") as sql_file:
    for bank_name in bank_names:
        icon_url = f"https://his-test1.oss-cn-hangzhou.aliyuncs.com/saas/bank/icon/icon/{bank_name}_ICON.png"
        logo_url = f"https://his-test1.oss-cn-hangzhou.aliyuncs.com/saas/bank/logo/bank_logo/{bank_name}_LOGO.png"
        sql = (
            f"INSERT INTO bank_icon (bank_name_en, icon_url, logo_url) "
            f"VALUES ('{bank_name}', '{icon_url}', '{logo_url}');\n"
        )
        sql_file.write(sql)

print("SQL 语句已成功写入 sql.sql 文件。")
```
